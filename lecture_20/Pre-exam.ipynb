### Задачка 1

Скачайте файл [languages.csv](https://storage.yandexcloud.net/pykili/19-2020/exam/language.zip) с выборкой языков из датасета WALS (World Atlas of Language Structures).

Это csv-таблица, каждая строка в ней соответствует какому-то языку.
Про каждый язык известно (колонки):

* wals_code	-- код языка в бд wals
* iso_code	-- код языка в стандарте ISO 639-3
* glottocode -- код языка в системе Glottolog
* **Name -- название языка (в Wals)**
* latitude -- широта для примерного геотега, соответствующего языку
* longitude -- долгота для примерного геотега, соответствующего языку
* **genus** -- genus, которому принадлежит язык (iirc это не очень однозначно переводится на русский, но если вы меня научите после экзамена, буду рад)
* **family** -- семья, котороый принадлежит язык
* **macroarea** -- макроареал, которому принадлежит язык
* countrycodes -- коды стран, в которых используется язык
* **1A Consonant Inventories; 2A Vowel Quality Inventories; ... -- признаки языков ([ссылка на подробное описание признаков](https://wals.info/feature))**

Иногда у какого-то языка ячейка, соответствующая признаку, пуста. В таком случае считаем, что значение признака неизвестно для этого языка.

Выделите две группы признаков:

1. Значения которых известны для большинства языков
2. Значения которых неизвестны для большинства языков

Сохраните признаки из первой группы в файл `wellknown_features.txt`, признаки из второй группы, соответствующие паттерну "... in ... [lL]anguages..." - в файл `specific_features.txt`, остальные признаки из второй группы - в файл `rare_features.txt`

В русском жестовом языке есть такие особенные штуки, которые называются классификаторами. Это такие жесты, которые не являются частью фиксированного лексикона, создаются на лету и обозначают какой-то предмет или движение кого-то или чего-то в пространстве. 

Дан файл с набором таких конструкций из корпуса:

```
CLF:тонкий.маленький.предмет:держать
CLF:прямоугольная.форма
CLF:человек.двигаться
CLF:машина
```

Базовое обозначение такое: `CLF:сущ:глагол`, то есть тег классифкатора, а потом кто делает и что делает. Такая конструкция тоже легальная: `CLF:сущ`. А вот это уже ошибка, если написано через точку: `CLF:сущ.глаг`, если к глаголу нет никакого существительного `CLF:глаг` -- это тоже ошибка.

1. Посчитайте сколько ошибок в данных в процентном соотношении и в абсолютных числах.
2. Сделайте json файл, в котором будет словарь, где ключ - это существительное, а значение - это список всех глаголов, которые с ним могут сочетаться.



### Задачка 3

Для работы понадобится файл с данными - [exam.zip (18MB)](https://storage.yandexcloud.net/pykili/exam.zip)

Архив содержит 7 больших файлов с текстами, обработанными через mystem.
Каждый большой файл - это склейка более маленьких файлов. Они собраны вместе для удобства транспортировки.

Каждый маленький файл имеет следующую структуру:

```xml
<html path="dir1/dir2">
<body>
...
</body>
</html>
```

### Задания

#### № 1

Напишите программу, которая разделит большие файлы на маленькие. Каждый маленький файл нужно положить в папку, в соответствии со значением в артибуте `path` тега `html`. Название файла выберите произвольное (уникальное). Расширение файла должно быть `.mystem`.

#### № 2

После разделения больших файлов у вас должны были получиться файлы со следующей структурой:

```xml
<html path="dir1/dir2">
<body>
<se>
<w><ana lex="..." gr="..."/><ana lex="..." gr="..."/>слово</w>
...
</se>
...
</body>
</html>
```

Каждое слово предложения имеет несколько вариантов морфологического разбора. Каждый из вариантов вместе со своей леммой содержится в теге `ana`. Самым первым идет наиболее вероятный разбор.

- извлеките из каждого файла `.mystem` все предложения и запишите каждое на новой строке рядом с исходным файлом в другом файле с расширением `.sent`. Предложения должны быть сформированы из **наименее** вероятных лемм.
- ... **наиболее** вероятных лемм
- ... только из существительных
